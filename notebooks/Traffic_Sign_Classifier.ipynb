### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
buffer_size = int(1e4)
batch_size = 128

# create tensorflow datasets
train_ds = make_dataset.from_tensor_slices(X_train, y_train, batch_size, shuffle=True, buffer_size=buffer_size)
valid_ds = make_dataset.from_tensor_slices(X_valid, y_valid, batch_size)
test_ds = make_dataset.from_tensor_slices(X_test, y_test, batch_size)

# preprocess the data(sets) by normalization and grayscaling
train_ds_prepared = build_features.preprocess(train_ds)
valid_ds_prepared = build_features.preprocess(valid_ds)
test_ds_prepared = build_features.preprocess(test_ds)

# show me some
visualize.image_grid(train_ds, class_names, 'Original training images')
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'training_examples.png'))

# show me some
visualize.image_grid(train_ds_prepared, class_names, 'Preprocessed training images')
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'training_examples_preprocessed.png'))
````

### Model Architecture

````python
### Define your architecture here.

# get the imput shape from the an image from the training set
input_shape = build_features.get_input_shape(train_ds_prepared)

# get the LeNet model architecture
model = train_model.lenet_architecture(input_shape, n_classes)

# show it
model.summary()
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.

history = train_model.train(model, train_ds_prepared, valid_ds_prepared, n_train, batch_size)
````

````python
accuracy_threshold = 0.93

# visualise the history
visualize.history(history, accuracy_threshold)

# save the figure for the report 
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'training_history.png'))
````

````python
# create tensorflow training dataset without shuffling 
train_ds_unshuffled = make_dataset.from_tensor_slices(X_train, y_train, batch_size)
train_ds_unshuffled_prepared = build_features.preprocess(train_ds_unshuffled)

### Calculate and report the accuracy on the training and validation set.
train_loss, train_acc = model.evaluate(train_ds_unshuffled_prepared, verbose=0)
valid_loss, valid_acc = model.evaluate(valid_ds_prepared, verbose=0)

### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
test_loss, test_acc = model.evaluate(test_ds_prepared, verbose=0)

# collect the data in a dataframe
evaluations = pd.DataFrame([
    {'dataset':'training', 'accuracy':train_acc, 'loss':train_loss},
    {'dataset':'validation', 'accuracy':valid_acc, 'loss':valid_loss},
    {'dataset':'testing', 'accuracy':test_acc, 'loss':test_loss}
]).set_index('dataset').sort_values('accuracy', ascending=True)
    
# visualise the evaluations
visualize.evaluations(evaluations, accuracy_threshold)

# save the figure for the report 
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'model_accuracy_on_datasets.png'))
````

## Step 3: Test a Model on New Images

To give yourself more insight into how your model is working, download at least five pictures of German traffic signs from the web and use your model to predict the traffic sign type.

You may find `signnames.csv` useful as it contains mappings from the class id (integer) to the actual sign name.

### Load and Output the Images

````python
### Load the images and plot them here.
custom_ds = tf.keras.preprocessing.image_dataset_from_directory(
    os.path.join(os.pardir, 'data', 'processed', 'custom_images'),
    shuffle=False,
    image_size=(32, 32),
    batch_size=batch_size
)
# get the custom class names
custom_class_names = custom_ds.class_names

# cast images from floats to ints
custom_ds = custom_ds.map(lambda image, label: (tf.cast(image, tf.int32), label), num_parallel_calls=tf.data.AUTOTUNE)

# show me some
visualize.image_grid(custom_ds, class_names, 'Original custom images')

# save the figure for the report         
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'custom_images.png'))
````

### Predict the Sign Type for Each Image

````python
### Run the predictions here and use the model to output the prediction for each image.
### Make sure to pre-process the images with the same pre-processing pipeline used earlier.

# preprocess the images
custom_ds_prepared = build_features.preprocess(custom_ds)

# get the actual value
_, labels_batch = next(iter(custom_ds_prepared))
y_true = [class_names[label] for label in labels_batch.numpy()]

# make predictions
y_pred, probabilities = predict_model.dataset(model, custom_ds_prepared, class_names)
````

````python
# add a softmax layer to the model
probability_model = tf.keras.Sequential([model, tf.keras.layers.Softmax()])

# predict the class probability of the new images
predictions = probability_model.predict(custom_ds_prepared)

# collect the labels here
y_true = []
y_pred = []

# show the images with the corresponding true & predicted labels
plt.figure(figsize=(23, 7))
for images, labels in custom_ds.take(1):
    for i in range(11):
        true_class_id = int(custom_ds.class_names[labels[i].numpy()])
        pred_class_id = int(np.argmax(predictions[i]))
        
        y_true.append(true_class_id)
        y_pred.append(pred_class_id)
        
        # green = predicted correctly, red otherwise
        color = 'green' if true_class_id == pred_class_id else 'red'
        
        ax = plt.subplot(2, 6, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))

        plt.title(class_names[true_class_id])
        plt.ylabel(class_names[pred_class_id], color=color)
        plt.axis("off")

# save the figure for the report         
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'custom_images_true_vs_pred.png'))     
````

### Analyze Performance

````python
### Calculate the accuracy for these 5 new images. 
### For example, if the model predicted 1 out of 5 signs correctly, it's 20% accurate on these new images.

# get a classification report
report = classification_report(y_true, y_pred, output_dict=True, zero_division=0)

# collect the data in a dataframe
evaluations = pd.DataFrame([
    {'dataset':'training', 'accuracy':train_acc, 'loss':train_loss},
    {'dataset':'validation', 'accuracy':valid_acc, 'loss':valid_loss},
    {'dataset':'testing', 'accuracy':test_acc, 'loss':test_loss},
    {'dataset':'custom', 'accuracy':report["accuracy"], 'loss':None},
]).set_index('dataset').sort_values('accuracy', ascending=True)
    
# visualise the evaluations
visualize.evaluations(evaluations, accuracy_threshold)

# save the figure for the report 
plt.savefig(os.path.join(os.pardir, 'reports', 'figures', 'model_accuracy_on_datasets_and_custom.png'))
````

````python
### Print out the top five softmax probabilities for the predictions on the German traffic sign images found on the web. 
probabilities, class_ids = tf.nn.top_k(predictions, k=5)

# loop over the images
for i, _ in enumerate(probabilities):
    
    # print the image id + ground thruth
    print('\n' + '-' * 20 + f'< Image: {i} >' + '-' * 20)
    print(f'ground thruth\t\t= {class_names[y_true[i]]} \n---')
    
    # loop over the top k predictions for the image
    for j, _ in enumerate(probabilities[i]):
        
        # print each class + probability
        print(f'{probabilities[i][j].numpy() * 100:.4f}% probability\t@ {class_names[class_ids[i][j].numpy()]}')
````
